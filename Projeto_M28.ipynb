## 2\. Análise Exploratória de Dados

Nesta análise iremos usar o seguinte pacote de dados:

```
import math
from typing import Iterator
from datetime import datetime, timedelta

import numpy as np
import pandas as pd
```

### **2\.1. Casos - Brasil**

Primeiramente, iremos processar os dados de **casos** da universidade de John Hopkins.

#### **2.1.1. Extração**

Os dados estão compilados em um arquivo por dia. Portanto, precisaremos extrair os dias necessários para a análise, que será o ano de 2021 inteiro.

E então selecionar as colunas de interesse e as linhas referentes ao Brasil, pois vamos trabalhar em cima apenas dos dados brasileiros.

```
# Exemplo usando o dado de 01/12/2021.
cases = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01-12-2021.csv', sep=',')
```

```
cases.head(5000)
```

```
def date_range(start_date: datetime, end_date: datetime) -> Iterator[datetime]:
  date_range_days: int = (end_date - start_date).days
  for lag in range(date_range_days + 1):
    yield start_date + timedelta(lag)
```

```
start_date = datetime(2021,  1,  1)
end_date   = datetime(2021, 12, 31)
```

```
cases = None
cases_is_empty = True

for date in date_range(start_date=start_date, end_date=end_date):

  date_str = date.strftime('%m-%d-%Y')
  data_source_url = f'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/{date_str}.csv'

  case = pd.read_csv(data_source_url, sep=',')

  case = case.drop(['FIPS', 'Admin2', 'Last_Update', 'Lat', 'Long_', 'Recovered', 'Active', 'Combined_Key', 'Case_Fatality_Ratio'], axis=1)
  case = case.query('Country_Region == "Brazil"').reset_index(drop=True)
  case['Date'] = pd.to_datetime(date.strftime('%Y-%m-%d'))
  if cases_is_empty:
    cases = case
    cases_is_empty = False
  else:
    cases = pd.concat([cases, case], axis= 0, ignore_index=True)
```

#### **2.1.2. Wrangling**

Aqui, faremos o ajuste dos dados para que fique melhor apresentável.

```
cases.head()
```

```
cases.info()
```

Vamos começar ajustando o nome das colunas, e logo a seguir, o nome dos estados.

```
cases = cases.rename(
  columns={
    'Province_State': 'state',
    'Country_Region': 'country'
  }
)

for col in cases.columns:
  cases = cases.rename(columns={col: col.lower()})
```

```
states_map = {
    'Amapa': 'Amapá',
    'Ceara': 'Ceará',
    'Espirito Santo': 'Espírito Santo',
    'Goias': 'Goiás',
    'Para': 'Pará',
    'Paraiba': 'Paraíba',
    'Parana': 'Paraná',
    'Piaui': 'Piauí',
    'Rondonia': 'Rondônia',
    'Sao Paulo': 'São Paulo'
}

cases['state'] = cases['state'].apply(lambda state: states_map.get(state) if state in states_map.keys() else state)
```

```
cases.head()
```

A seguir, adicionaremos mais algumas colunas a fim de enriquecer nossa base de dados.

 - Chaves temporais:

```
cases['month'] = cases['date'].apply(lambda date: date.strftime('%m'))
cases['year']  = cases['date'].apply(lambda date: date.strftime('%Y'))
```

 - População estimada do estado:

```
cases['population'] = round(100000 * (cases['confirmed'] / cases['incident_rate']))
cases = cases.drop('incident_rate', axis=1)
```

- Número, média móvel (7 dias) e estabilidade (14 dias) de casos confirmados e mortes por estado:

```
cases_ = None
cases_is_empty = True

def get_trend(rate: float) -> str:

  if np.isnan(rate):
    return np.nan

  if rate < 0.75:    # Se a divisão for menor que 0.75 (correção 0.85)
    status = 'downward'
  elif rate > 1.15:
    status = 'upward'
  else:
    status = 'stable'

  return status


for state in cases['state'].drop_duplicates():

  cases_per_state = cases.query(f'state == "{state}"').reset_index(drop=True)
  cases_per_state = cases_per_state.sort_values(by=['date'])

  cases_per_state['confirmed_1d'] = cases_per_state['confirmed'].diff(periods=1)
  cases_per_state['confirmed_moving_avg_7d'] = np.ceil(cases_per_state['confirmed_1d'].rolling(window=7).mean())
  cases_per_state['confirmed_moving_avg_7d_rate_14d'] = cases_per_state['confirmed_moving_avg_7d']/cases_per_state['confirmed_moving_avg_7d'].shift(periods=14)
  cases_per_state['confirmed_trend'] = cases_per_state['confirmed_moving_avg_7d_rate_14d'].apply(get_trend)

  cases_per_state['deaths_1d'] = cases_per_state['deaths'].diff(periods=1)
  cases_per_state['deaths_moving_avg_7d'] = np.ceil(cases_per_state['deaths_1d'].rolling(window=7).mean())
  cases_per_state['deaths_moving_avg_7d_rate_14d'] = cases_per_state['deaths_moving_avg_7d']/cases_per_state['deaths_moving_avg_7d'].shift(periods=14)
  cases_per_state['deaths_trend'] = cases_per_state['deaths_moving_avg_7d_rate_14d'].apply(get_trend)

  if cases_is_empty:
    cases_ = cases_per_state
    cases_is_empty = False
  else:
    cases_ = pd.concat([cases_, cases_per_state],axis=0, ignore_index=True)

cases = cases_
cases_ = None
```

E então, faremos o *type casting* das colunas, garantindo assim, uma base mais sólida dos dados.

```
cases['population'] = cases['population'].astype('Int64')
cases['confirmed_1d'] = cases['confirmed_1d'].astype('Int64')
cases['confirmed_moving_avg_7d'] = cases['confirmed_moving_avg_7d'].astype('Int64')
cases['deaths_1d'] = cases['deaths_1d'].astype('Int64')
cases['deaths_moving_avg_7d'] = cases['deaths_moving_avg_7d'].astype('Int64')
```

Por fim, reorganizaremos as colunas com o intuito de ficar melhor apresentável.

```
cases = cases[['date', 'country', 'state', 'population', 'confirmed', 'confirmed_1d', 'confirmed_moving_avg_7d', 'confirmed_moving_avg_7d_rate_14d', 'confirmed_trend', 'deaths', 'deaths_1d', 'deaths_moving_avg_7d', 'deaths_moving_avg_7d_rate_14d', 'deaths_trend', 'month', 'year']]
```

### **2.2. Vacinação - Brasil**

Agora, processaremos os dados de vacinação da Universidade de Oxford.

```
vaccines = vaccines.query('location == "Brazil"').reset_index(drop=True)
vaccines = vaccines[['location', 'population', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'total_boosters', 'date']]
```

#### **2.2.2. Wrangling**

Agora manipularemos os dados para o Dashboard. Garantindo assim, uma boa granularidade da base de dados.

```
vaccines.info()
```

```
vaccines = vaccines.ffill()
```

E então, filtraremos os dados para que sejam a mesma data contida nos casos, e alteraremos o nome das colunas.

```
vaccines = vaccines[(vaccines['date'] >= '2021-01-01') & (vaccines['date'] <= '2021-12-31')].reset_index(drop=True)
```

```
vaccines = vaccines.rename(
  columns={
    'location': 'country',
    'total_vaccinations': 'total',
    'people_vaccinated': 'one_shot',
    'people_fully_vaccinated': 'two_shots',
    'total_boosters': 'three_shots',
  }
)
```

Logo após, adicionaremos mais algumas informações a fim de enriquecer a base de dados, assim como nos Casos.

```
vaccines['month'] = vaccines['date'].apply(lambda date: date.strftime('%m'))
vaccines['year']  = vaccines['date'].apply(lambda date: date.strftime('%Y'))
```

 - Dados relativos:

```
vaccines['one_shot_perc'] = round(vaccines['one_shot'] / vaccines['population'], 4)
vaccines['two_shots_perc'] = round(vaccines['two_shots'] / vaccines['population'], 4)
vaccines['three_shots_perc'] = round(vaccines['three_shots'] / vaccines['population'], 4)
```

Por fim, garantiremos que o tipo de dado seja o correto, fazendo um *type casting* das colunas, e então as reorganizaremos.

```
vaccines['population'] = vaccines['population'].astype('Int64')
vaccines['total'] = vaccines['total'].astype('Int64')
vaccines['one_shot'] = vaccines['one_shot'].astype('Int64')
vaccines['two_shots'] = vaccines['two_shots'].astype('Int64')
vaccines['three_shots'] = vaccines['three_shots'].astype('Int64')

vaccines = vaccines[['date', 'country', 'population', 'total', 'one_shot', 'one_shot_perc', 'two_shots', 'two_shots_perc', 'three_shots', 'three_shots_perc', 'month', 'year']]
```

## **4\. Conclusão**

Comparando os dois gráficos, podemos observar diferenças significativas entre o Brasil e o Canadá ao longo do ano de 2021.

O Brasil teve os números mais inflados que o Canadá em todos os aspectos. Até o mês de Junho, o número de casos no Brasil foi extremamente alto, então decaindo lentamente, demonstrando uma melhora no controle dos casos obtidos. Por outro lado, no Canadá, o número de casos ao longo do ano foram bem mais controlados, com uma explosão intensa no final do ano, provavelmente devido a variante Ômicron.

Ao longo do ano, o Brasil teve, também, um número de mortes muito maior que o Canadá, entretanto seu pico de mortes fora diferente em cada país, o Brasil sendo mais ou menos em abril e no Canadá no começo do ano. E, é claro, a curva de mortes decaiu mais rapidamente no Canadá, demonstrando assim, um melhor controle de danos e eficácia nas medidas de mitigação.

O Brasil por um lado, enfrentou maiores dificuldades em relação a pandemia ao longo do ano, provavelmente devido a medidas de contenção menos rígidas, ausência de responsabilidade de muitas pessoas e dificuldade em estruturas que o país passa, transporte público lotados, entre outros. Também se da por, provavelmente, muitas pessoas precisarem sair de casa para conseguir botar comida na mesa no fim do dia, não havendo a possibilidade de trabalhar remotamente.

Por outro lado, o Canadá, passou por menos dificuldades que o Brasil, conseguindo manter todos os números controlados, tanto de casos, quanto de mortes e vacinação. Apesar de seu grande pico no final do ano, as vacinações conseguiram mitigar bem a situação, mantendo um número de mortes mais baixa.

Por fim, devemos levar em consideração todas as festas em datas comemorativas, como Natal e Ano novo e a variante Ômicron.

Logo, podemos concluir que o Brasil passou por muito mais dificuldades que o Canadá ao longo de 2021, especialmente no primeiro semestre. O Canadá no geral, conseguiu manter os números mais controlados que o Brasil, apesar da explosão no fim do ano. Sugerindo assim que políticas de saúde pública, medidas de contenção e vacinação, desempenharam um papel crucial no impacto da pandemia de cada país, porém, com a falta de estrutura, medidas de contenção menos rígidas e pessoas irresponsáveis, o Brasil teve, infelizmente, uma taxa de casos e mortes muito maiores.
